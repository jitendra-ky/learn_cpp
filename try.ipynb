```python
import csv

name = []

with open('cities_all.csv', mode='r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    
    # Optionally, you can get the header
    header = next(csv_reader)
    print(f'Header: {header}')

    # Print each row
    for row in csv_reader:
        if row[3] == 'IN':
            name.append(row[1])

print(name)
```
